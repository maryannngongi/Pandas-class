```python
#striping whitespace from column names
data.columns = data.columns.str.strip()
data.columns
```

```python
#rename columns
data = data.rename(columns = {'humidity':'avg_humidity'})
data.columns
```

```python
data.info()
```

```python
#type to integer
data['meanpressure.int'] = data['meanpressure'].astype('int') 
data.info()
```

```python
#converting date column to datetime format
data['date'] = pd.to_datetime(data['date'])
data.info()
```

```python
def categorize_temp(temp):
    if temp < 20:
        return 'cold'
    elif temp >= 20:
            return 'warm'
    else:
                return 'hot'
```

```python
data['temperature_category'] = data['meantemp'].apply(categorize_temp)
```

```python
#.map()
#month
data['month'] = data['date'].dt.month
data
```

```python
#mapping
month_map = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr',5:'May',6:'June', 7:'July', 8:'Aug', 9:'Sept', 10:'Oct', 11:'Nov', 12:'Dec'}
data['month_name'] = data['month'].map(month_map)
data
```

```python
data['day_of_week'] = data['date'].dt.dayofweek
data
```

```python
day_of_week_map = {0:'Mon',1:'Tue',2:'Wed',3:'Thur',4:'Fri',5:'sat',6:'sun'}
data['day_of_week_name'] = data['day_of_week'].map(day_of_week_map)
data
```
